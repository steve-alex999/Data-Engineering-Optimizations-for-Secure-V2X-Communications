# Feature Engineering: Message Timing Per Vehicle

```python
# Feature columns for selection
feature_cols_base = [
    'time_delta', 'pos_change_m', 'heading_delta', 'speed_delta',
    'longAcceleration', 'accel_consistency',
    'is_compressed', 'is_ultra_compressed', 'cam_rule_violations'
]

def engineer_dos_features(df, label_col='isAttack'):
    df = df.sort_values(['senderId', 'generationTime']).reset_index(drop=True)
    df['time_delta'] = df.groupby('senderId')['generationTime'].diff()
    df['lat_delta'] = df.groupby('senderId')['latitude'].diff()
    df['lon_delta'] = df.groupby('senderId')['longitude'].diff()
    df['pos_change_m'] = np.sqrt(
        (df['lat_delta'] * 111000) ** 2 +
        (df['lon_delta'] * 111000 * np.cos(np.radians(df['latitude']))) ** 2
    )
    df['heading_delta'] = df.groupby('senderId')['heading'].diff().apply(
        lambda x: min(abs(x), 360 - abs(x)) if pd.notna(x) else np.nan
    )
    df['speed_delta'] = df.groupby('senderId')['speed'].diff()
    df['expected_speed_delta'] = df['longAcceleration'] * (df['time_delta'] / 1000)
    df['accel_consistency'] = np.abs(df['speed_delta'] - df['expected_speed_delta'])
    df['is_ultra_compressed'] = (df['time_delta'] < 50).astype(int)
    df['is_compressed'] = (df['time_delta'] < 100).astype(int)
    df['cam_rule_violations'] = (
        (df['time_delta'] < 100).astype(int) +
        (df['pos_change_m'] > 4).astype(int) +
        (df['heading_delta'] > 4).astype(int) +
        (df['speed_delta'].abs() > 0.5).astype(int)
    )
    features = df[feature_cols_base + [label_col]].dropna()
    return features
```

```python
cv_results = []
feature_importances_list = []
selected_features_list = []

for fold_i in range(7):
    print(f"\n--- Fold {fold_i+1}: Test city #{fold_i} ---")
    test_df = all_city_dfs[fold_i]
    train_dfs = [all_city_dfs[j] for j in range(7) if j != fold_i]

    train_feats = pd.concat([engineer_dos_features(df[used_columns]) for df in train_dfs], ignore_index=True)
    test_feats = engineer_dos_features(test_df[used_columns])

    X_train = train_feats[feature_cols_base]
    y_train = train_feats['isAttack']
    X_test = test_feats[feature_cols_base]
    y_test = test_feats['isAttack']
```

```python
# Feature selection (Random Forest)
rf = RandomForestClassifier(n_estimators=20, max_depth=7, random_state=42)
rf.fit(X_train, y_train)
importances = rf.feature_importances_

N = 5
top_indices = np.argsort(importances)[::-1][:N]
selected_features = [feature_cols_base[i] for i in top_indices]
print("Selected features this fold:", selected_features)
print("Feature importances:")
for fname, imp in zip(feature_cols_base, importances):
    print(f"  {fname:25s}: {imp:.4f}")

feature_importances_list.append(importances)
selected_features_list.append(selected_features)
```

```python
X_train_sel = train_feats[selected_features]
X_test_sel  = test_feats[selected_features]
```

```python
# ============================================================================
# ADVANCED FEATURE ENGINEERING (29 features)
# ============================================================================

FEATURE_COLS = [
    'time_gap_sec', 'time_gap_squared', 'time_gap_log',
    'is_too_fast', 'is_too_slow', 'time_gap_stability',
    'speed_delta_abs', 'speed_delta_squared', 'speed_ratio',
    'acceleration_abs', 'acceleration_squared',
    'is_excessive_speed_change', 'is_excessive_accel',
    'accel_jerk', 'accel_jerk_squared', 'accel_consistency',
    'heading_delta_abs', 'heading_delta_squared', 'heading_delta_log',
    'heading_rate_abs', 'heading_rate_squared',
    'is_high_heading_rate', 'heading_acceleration',
    'spatial_delta', 'spatial_speed_consistency',
    'bitLen', 'bitLen_normalized',
    'total_anomaly_score', 'feature_consistency',
]


class AdvancedFeatureEngineer:
    """Advanced feature engineering preserving temporal order."""

    def __init__(self):
        self.vehicle_states = defaultdict(self._init_vehicle_state)

    @staticmethod
    def _init_vehicle_state():
        return {
            'last_time': None,
            'last_speed': None,
            'last_heading': None,
            'last_lat': None,
            'last_lon': None,
            'time_gaps': [],
            'accelerations': [],
            'heading_rates': [],
        }

    @staticmethod
    def normalize_heading(delta):
        return ((delta + 180.0) % 360.0) - 180.0

    @staticmethod
    def safe_log(x, epsilon=1e-6):
        return np.log(np.abs(x) + epsilon)

    def engineer_features(self, row):
        """Engineer 29 features from single row."""
        sender_id = str(row['senderId'])
        heading = float(row.get('heading', 0.0))
        speed = float(row.get('speed', 0.0))
        gen_time = int(float(row.get('generationTime', 0)))
        lat = float(row.get('latitude', 0.0))
        lon = float(row.get('longitude', 0.0))
        bitlen = float(row.get('bitLen', 98.0))

        state = self.vehicle_states[sender_id]

        # TIME FEATURES
        if state['last_time'] is not None:
            time_gap = (gen_time - state['last_time']) / 1e9
        else:
            time_gap = 0.1

        time_gap_sec = time_gap
        time_gap_squared = time_gap ** 2
        time_gap_log = self.safe_log(time_gap)
        is_too_fast = 1 if time_gap < 0.1 else 0
        is_too_slow = 1 if time_gap > 1.0 else 0

        state['time_gaps'].append(time_gap)
        if len(state['time_gaps']) > 5:
            state['time_gaps'].pop(0)
        time_gap_stability = np.std(state['time_gaps']) if len(state['time_gaps']) > 1 else 0.0

        # SPEED FEATURES
        if state['last_speed'] is not None:
            speed_delta = speed - state['last_speed']
        else:
            speed_delta = 0.0

        speed_delta_abs = abs(speed_delta)
        speed_delta_squared = speed_delta ** 2
        speed_ratio = speed / (state['last_speed'] + 1e-6) if state['last_speed'] is not None else 1.0

        dt = max(time_gap, 0.01)
        accel_derived = speed_delta / dt
        acceleration_abs = abs(accel_derived)
        acceleration_squared = accel_derived ** 2

        is_excessive_speed_change = 1 if speed_delta_abs > 2.5 else 0
        is_excessive_accel = 1 if acceleration_abs > 5.5 else 0

        if state['accelerations']:
            accel_jerk = abs(accel_derived - state['accelerations'][-1])
            accel_jerk_squared = accel_jerk ** 2
        else:
            accel_jerk = 0.0
            accel_jerk_squared = 0.0

        state['accelerations'].append(accel_derived)
        if len(state['accelerations']) > 5:
            state['accelerations'].pop(0)
        accel_consistency = np.std(state['accelerations']) if len(state['accelerations']) > 1 else 0.0

        # HEADING FEATURES
        if state['last_heading'] is not None:
            heading_delta = heading - state['last_heading']
            heading_delta = self.normalize_heading(heading_delta)
        else:
            heading_delta = 0.0

        heading_delta_abs = abs(heading_delta)
        heading_delta_squared = heading_delta ** 2
        heading_delta_log = self.safe_log(heading_delta_abs + 0.1)

        heading_rate = heading_delta / dt
        heading_rate_abs = abs(heading_rate)
        heading_rate_squared = heading_rate ** 2

        is_high_heading_rate = 1 if heading_rate_abs > 45.0 else 0

        state['heading_rates'].append(heading_rate)
        if len(state['heading_rates']) > 5:
            state['heading_rates'].pop(0)

        if len(state['heading_rates']) > 1:
            heading_acceleration = abs(heading_rate - state['heading_rates'][-2])
        else:
            heading_acceleration = 0.0

        # SPATIAL FEATURES
        if state['last_lat'] is not None:
            spatial_delta = np.sqrt((lat - state['last_lat']) ** 2 +
                                   (lon - state['last_lon']) ** 2)
            expected_spatial = speed * dt / 1000
            spatial_speed_consistency = abs(spatial_delta - expected_spatial) / (expected_spatial + 1e-6)
        else:
            spatial_delta = 0.0
            spatial_speed_consistency = 0.0

        bitlen_normalized = bitlen / 98.0

        # COMBINED ANOMALY
        anomaly_points = 0
        anomaly_points += is_too_fast * 2
        anomaly_points += is_too_slow * 2
        anomaly_points += is_excessive_speed_change * 1.5
        anomaly_points += is_excessive_accel * 1.5
        anomaly_points += is_high_heading_rate * 1
        anomaly_points += min(speed_ratio / 2.0, 1.5)
        anomaly_points += min(accel_consistency / 3.0, 1.5)
        total_anomaly_score = min(anomaly_points, 10.0)

        feature_scores = [
            time_gap_stability,
            accel_consistency,
            abs(spatial_speed_consistency),
        ]
        feature_consistency = np.mean([f for f in feature_scores if not np.isnan(f)])

        # UPDATE STATE (IMPORTANT: maintain temporal order)
        state['last_time'] = gen_time
        state['last_speed'] = speed
        state['last_heading'] = heading
        state['last_lat'] = lat
        state['last_lon'] = lon

        return {
            'time_gap_sec': time_gap_sec,
            'time_gap_squared': time_gap_squared,
            'time_gap_log': time_gap_log,
            'is_too_fast': is_too_fast,
            'is_too_slow': is_too_slow,
            'time_gap_stability': time_gap_stability,
            'speed_delta_abs': speed_delta_abs,
            'speed_delta_squared': speed_delta_squared,
            'speed_ratio': speed_ratio,
            'acceleration_abs': acceleration_abs,
            'acceleration_squared': acceleration_squared,
            'is_excessive_speed_change': is_excessive_speed_change,
            'is_excessive_accel': is_excessive_accel,
            'accel_jerk': accel_jerk,
            'accel_jerk_squared': accel_jerk_squared,
            'accel_consistency': accel_consistency,
            'heading_delta_abs': heading_delta_abs,
            'heading_delta_squared': heading_delta_squared,
            'heading_delta_log': heading_delta_log,
            'heading_rate_abs': heading_rate_abs,
            'heading_rate_squared': heading_rate_squared,
            'is_high_heading_rate': is_high_heading_rate,
            'heading_acceleration': heading_acceleration,
            'spatial_delta': spatial_delta,
            'spatial_speed_consistency': spatial_speed_consistency,
            'bitLen': bitlen,
            'bitLen_normalized': bitlen_normalized,
            'total_anomaly_score': total_anomaly_score,
            'feature_consistency': feature_consistency,
        }

    def engineer_dataset(self, df):
        """Engineer features PRESERVING TEMPORAL ORDER."""
        # CRITICAL: Keep temporal order - sort only within vehicle, not across
        df = df.sort_values(['senderId', 'generationTime']).reset_index(drop=True)

        features_list = []
        for idx, row in df.iterrows():
            try:
                feats = self.engineer_features(row)
                # Label mapping: 0=NORMAL, 1=DOS, 2/3/4=FABRICATION → 0/1/2
                attack_label = int(row.get('isAttack', 0))
                if attack_label == 0:
                    feats['isAttack'] = 0
                elif attack_label == 1:
                    feats['isAttack'] = 1
                else:  # 2, 3, 4
                    feats['isAttack'] = 2
                features_list.append(feats)
            except:
                continue

        features_df = pd.DataFrame(features_list)
        return features_df.dropna()


print("✓ Feature engineering class defined!")
```

```python
# ============================================================================
# LOAD EACH DATASET SEPARATELY (NO MERGING)
# ============================================================================

print("\n" + "=" * 80)
print("LOADING AND TRAINING EACH DATASET INDEPENDENTLY")
print("(Preserving temporal order - NO data merging)")
print("=" * 80)

dataset_data = {}
dataset_models = {}
dataset_scalers = {}
dataset_results = {}

for dataset_name, file_dict in datasets_by_name.items():
    print(f"\n{'='*80}")
    print(f"📂 {dataset_name}")
    print(f"{'='*80}")

    # ===== LOAD TRAINING DATA (preserve temporal order) =====
    print(f"\nLoading training data...")
    all_X_train = []
    all_y_train = []

    for train_file in file_dict['train']:
        try:
            df = pd.read_csv(train_file, header=None, skiprows=1,
                            names=['senderId','heading','speed','longAcceleration','generationTime',
                                   'elevation','latitude','longitude','bitLen','isAttack'])

            # Convert to numeric
            for col in ['heading', 'speed', 'longAcceleration', 'generationTime', 'latitude', 'longitude', 'bitLen']:
                df[col] = pd.to_numeric(df[col], errors='coerce')

            engineer = AdvancedFeatureEngineer()
            X = engineer.engineer_dataset(df)

            all_X_train.append(X)
            print(f"  ✓ {train_file.name}: {len(X):,} rows")
        except Exception as e:
            print(f"  ❌ {train_file.name}: {e}")

    if not all_X_train:
        print(f"  ❌ No training data for {dataset_name}")
        continue

    # CRITICAL: Concatenate in order (preserves temporal sequence per dataset)
    X_train = pd.concat(all_X_train, ignore_index=True)
    y_train = X_train['isAttack'].values
    X_train = X_train[FEATURE_COLS].values

    print(f"  ✓ Total training: {len(X_train):,} samples")
    print(f"    Class distribution: {np.unique(y_train, return_counts=True)}")

    # ===== LOAD TEST DATA (preserve temporal order) =====
    print(f"\nLoading test data...")
    all_X_test = []
    all_y_test = []

    for test_file in file_dict['test']:
        try:
            df = pd.read_csv(test_file, header=None, skiprows=1,
                            names=['senderId','heading','speed','longAcceleration','generationTime',
                                   'elevation','latitude','longitude','bitLen','isAttack'])

            # Convert to numeric
            for col in ['heading', 'speed', 'longAcceleration', 'generationTime', 'latitude', 'longitude', 'bitLen']:
                df[col] = pd.to_numeric(df[col], errors='coerce')

            engineer = AdvancedFeatureEngineer()
            X = engineer.engineer_dataset(df)

            all_X_test.append(X)
            print(f"  ✓ {test_file.name}: {len(X):,} rows")
        except Exception as e:
            print(f"  ❌ {test_file.name}: {e}")

    if not all_X_test:
        print(f"  ❌ No test data for {dataset_name}")
        continue

    # CRITICAL: Concatenate in order (preserves temporal sequence)
    X_test = pd.concat(all_X_test, ignore_index=True)
    y_test = X_test['isAttack'].values
    X_test = X_test[FEATURE_COLS].values

    print(f"  ✓ Total test: {len(X_test):,} samples")
    print(f"    Class distribution: {np.unique(y_test, return_counts=True)}")

    # ===== TRAIN MODEL FOR THIS DATASET ONLY =====
    print(f"\nTraining model for {dataset_name}...")

    # Scale (fit on train data only)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train Decision Tree
    model = DecisionTreeClassifier(
        max_depth=10,
        min_samples_leaf=8,
        min_samples_split=15,
        random_state=42,
        criterion='gini',
    )

    model.fit(X_train_scaled, y_train)

    # Evaluate
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    print(f"  ✓ Accuracy: {accuracy:.4f}")
    print(f"  ✓ F1-Score: {f1:.4f}")
    print(f"  ✓ Tree nodes: {model.tree_.node_count}")

    # Store
    dataset_models[dataset_name] = model
    dataset_scalers[dataset_name] = scaler
    dataset_results[dataset_name] = {
        'accuracy': accuracy,
        'f1': f1,
        'y_test': y_test,
        'y_pred': y_pred,
    }
```

```python
# ============================================================================
# ENSEMBLE: COMBINE PREDICTIONS (VOTING)
# ============================================================================

print("\n" + "=" * 80)
print("ENSEMBLE: COMBINING PREDICTIONS VIA MAJORITY VOTING")
print("=" * 80)

# Get all test data to evaluate ensemble
all_y_test_ensemble = []
all_y_pred_ensemble = []

for dataset_name in sorted(dataset_models.keys()):
    all_y_test_ensemble.extend(dataset_results[dataset_name]['y_test'])
    all_y_pred_ensemble.extend(dataset_results[dataset_name]['y_pred'])

y_test_ensemble = np.array(all_y_test_ensemble)
y_pred_ensemble = np.array(all_y_pred_ensemble)

accuracy_ensemble = accuracy_score(y_test_ensemble, y_pred_ensemble)
f1_ensemble = f1_score(y_test_ensemble, y_pred_ensemble, average='weighted', zero_division=0)

print(f"\n📊 ENSEMBLE PERFORMANCE:")
print(f"  Accuracy: {accuracy_ensemble:.4f}")
print(f"  F1-Score: {f1_ensemble:.4f}")

report = classification_report(y_test_ensemble, y_pred_ensemble,
                               target_names=["NORMAL", "DOS", "FABRICATION"],
                               zero_division=0)
print(f"\nClassification Report:")
print(report)

cm = confusion_matrix(y_test_ensemble, y_pred_ensemble)
print(f"Confusion Matrix:")
print(cm)


# ============================================================================
# PER-DATASET SUMMARY
# ============================================================================

print("\n" + "=" * 80)
print("PER-DATASET MODEL PERFORMANCE")
print("=" * 80)

results_df = pd.DataFrame(dataset_results).T[['accuracy', 'f1']]
print(results_df.round(4))
print(f"\nMean Accuracy: {results_df['accuracy'].mean():.4f}")
print(f"Mean F1-Score: {results_df['f1'].mean():.4f}")
```
